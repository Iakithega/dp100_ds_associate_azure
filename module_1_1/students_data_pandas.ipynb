```python
url = "https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/grades.csv"
response = requests.get(url)

with open("grades.csv", "wb") as f:
    f.write(response.content)
    
df_students = pd.read_csv('grades.csv',delimiter=',',header='infer')
df_students.head()
```

```python
df_students.loc[df_students['StudyHours'] > df_students['StudyHours'].mean()*1.15].mean()
```

```python
df_students['Passe'] = df_students['Grade'] >= 60
df_students = df_students.assign(Passas = df_students['Grade'] >= 60)

df_students
```

```python
df_students.groupby(df_students.Passe).count()
```

```python
print(df_students.groupby(df_students.Passe)[['StudyHours', 'Grade']].mean())
```
